```python
def abs_sobel_thresh(img, orient='x', thresh_min=10, thresh_max=255):
    # Grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    if 'x' in orient.lower():
        sobel = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
    elif 'y' in orient.lower():  
        sobel = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
    
    sobel = np.absolute(sobel)
    sobel -= sobel.min()
    sobel /= sobel.max()
    sobel = (sobel * 255).astype(np.uint8)

    bw = np.zeros_like(sobel)
    bw[(sobel >= thresh_min) & (sobel <= thresh_max)] = 1
    return bw
```

```python
def mag_thresh(img, sobel_kernel=3, mag_thresh=(0, 255)):
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    sobel = np.sqrt(np.square(sobelx) + np.square(sobely))
    sobel -= sobel.min()
    sobel /= sobel.max()
    sobel = (sobel * 255).astype(np.uint8)
    bw = np.zeros_like(sobel)
    bw[(sobel >= mag_thresh[0]) & (sobel <= mag_thresh[1])] = 1
    return bw
    
```

```python
def dir_threshold(img, sobel_kernel=3, thresh=(0, np.pi/2)):
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)

    sobel = np.arctan2(sobely, sobelx)
    
    sobel -= sobel.min()
    sobel /= sobel.max()
    sobel = (sobel * 255).astype(np.uint8)
    bw = np.zeros_like(sobel)
    bw[(sobel >= mag_thresh[0]) & (sobel <= mag_thresh[1])] = 1
    return bw
    
    
```
